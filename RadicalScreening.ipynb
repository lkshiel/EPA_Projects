```python
import rdkit
from rdkit import Chem
import rdkit.Chem.rdMolDescriptors
from rdkit.Chem import Descriptors
from rdkit.Chem.MolStandardize import rdMolStandardize
enumerator = rdMolStandardize.TautomerEnumerator()
```

```python
smiles=['CCO[O]','[CH2]C','C[C]','C[N]C','CC[N+]C','CC[NH2+]C','CCOO']
for smi in smiles:
    mol=Chem.MolFromSmiles(smi) 
    radical=rdkit.Chem.Descriptors.NumRadicalElectrons(mol) #get the number of radical electrons for chemical
    print(smi,'has',radical, 'radical electron(s)',)
    if radical <1: #chemicals with 1 or more radical electron are Radical chemicals and cannot be processed by CTS
        display(mol)
    else:
        print('The SMILES you entered is a radical and cannot be processed')
```
